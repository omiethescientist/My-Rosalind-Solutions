```python
def stringBinarySearch(strs, l, r, x):
  mid = l + (r - l) // 2
  if strs[mid:mid+2] == x:
    return mid+1
  elif strs[mid:mid+2] < x:
    return stringBinarySearch(strs,mid,r,x)
  else:
    return stringBinarySearch(strs,l,mid,x)
```

```python
def CountNucs(string):
  string = string.upper()
  sorted_characters = sorted(string)
  sorted_string = "".join(sorted_characters)
  end_idxs = {i[0]:stringBinarySearch(sorted_string, 0, len(sorted_string), i) for i in ["AC", "CG", "GT"]}
  counter = {"A":end_idxs["A"], "C":end_idxs["C"] - end_idxs["A"], "G":end_idxs["G"] - end_idxs["C"], "T":len(string) - end_idxs["G"]}
  return counter
```
