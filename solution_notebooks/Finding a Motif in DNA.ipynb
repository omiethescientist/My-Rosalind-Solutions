```python
motif = "CAAAGATCA"
```

```python
def argsort(seq):
  return sorted(range(len(seq)), key = seq.__getitem__)
```

```python
def motifSearch(string, l, r, motif):
  suffixes = [string[i:] for i in range(len(string))]
  suffixes = argsort(suffixes)
  if r < l:
    print("r must be less than l")
  mid = l + (r - l) // 2
  mid_suffix = string[suffixes[mid]:]
  if ((mid_suffix > motif) & (mid_suffix.startswith(motif) == False)):
    return motifSearch(string, l, mid-1, motif)
  elif mid_suffix < motif:
    return motifSearch(string, mid+1, r, motif)
  else:
    return [idx+1 for idx in suffixes[l:r] if string[idx:].startswith(motif)]
```

```python
print(*sorted(motifSearch(string, 0, len(string), motif)), sep = ' ')
```
